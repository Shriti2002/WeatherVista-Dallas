#**# CONTINGENCY TABLE : CONDITIONS VS ICONS**

# What is the relationship between weather conditions and icons, and is there a significant association between them?

*The heatmap visualizes the contingency table of weather conditions versus icons, showing the frequency of each combination. For example, "Partially cloudy" is most associated with the "partly-cloudy-day" icon (148 occurrences), while "Clear" aligns with "clear-day" (98 occurrences). The Chi-squared test statistic (1464.00) and p-value (0.0000) indicate a strong, statistically significant association between conditions and icons.*

```python
from scipy.stats import chi2_contingency

# Example: 'conditions' vs 'icon'
contingency_table = pd.crosstab(df['conditions'], df['icon'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-squared test statistic: {chi2:.2f}")
print(f"P-value: {p:.5f}")

# Optional: Visualize the table
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Contingency Table: Conditions vs Icons")
plt.xlabel("Icon")
plt.ylabel("Condition")
plt.tight_layout()
plt.show()
```

#**# TEMPERATURE OUTLIER DETECTION AND REMOVAL (Z- SCORE METHOD)**

# How are temperature outliers detected and removed using the Z-score method?

*The boxplot shows the distribution of temperature after removing outliers based on a Z-score threshold of 3. A total of 3 outliers were detected and excluded from the dataset. The cleaned data now reflects a more accurate representation of temperature without extreme values, as evident in the compact range of the boxplot. This process ensures robust analysis by mitigating the impact of outliers.*

```python
from scipy.stats import zscore

# Use 'temp' feature
df['temp_z'] = zscore(df['temp'])

# Define outliers
outliers = df[np.abs(df['temp_z']) > 3]

print("Outliers detected:", outliers.shape[0])
df_clean = df[np.abs(df['temp_z']) <= 3]

# Plot
sns.boxplot(x=df_clean['temp'])
plt.title('Temperature (Z-score Outlier Removal)')
plt.show()
```

#**# HUMIDITY OUTLIER DETECTION AND REMOVAL (ISOLATION FOREST)**

# How are outliers in humidity identified and removed using the Isolation Forest method?

*The boxplot shows the distribution of humidity after cleaning the data using the Isolation Forest algorithm with a contamination rate of 0.02. Outliers were successfully removed, resulting in a more compact and representative range of humidity values (approximately 35%90%). This process improves data quality by eliminating anomalies.*

```python
from sklearn.ensemble import IsolationForest

features = ['humidity', 'dew', 'uvindex']
clf = IsolationForest(contamination=0.02, random_state=42)
df['anomaly'] = clf.fit_predict(df[features])

# Keep only inliers
df_if_cleaned = df[df['anomaly'] == 1]

# Visualize one feature
sns.boxplot(x='humidity', data=df_if_cleaned)
plt.title('Humidity (Isolation Forest Cleaned)')
plt.show()
```

#**# SOLAR ENERGY OUTLIER DETECTION AND REMOVAL (MODIFIED Z-SCORE)**

# How are outliers in solar energy identified and removed using the Modified Z-Score method?

*The boxplot shows the distribution of solar energy after cleaning the data using the Modified Z-Score method with a threshold of 3.5. Outliers were successfully removed, resulting in a more compact and representative range of solar energy values (approximately 530). This ensures more reliable analysis by eliminating extreme values.*

```python
def modified_z_score(series):
    median = np.median(series)
    mad = np.median(np.abs(series - median))
    return 0.6745 * (series - median) / mad

df['solarenergy_mz'] = modified_z_score(df['solarenergy'])
df_mz_clean = df[np.abs(df['solarenergy_mz']) <= 3.5]

# Visualize
sns.boxplot(x='solarenergy', data=df_mz_clean)
plt.title('Solar Energy (Modified Z-score Cleaned)')
plt.show()
```

#**# DBSCAN CLEANED : TEMP VS HUMIDITY**

# How does the relationship between temperature and humidity appear after cleaning outliers using DBSCAN?

*The scatter plot shows the cleaned relationship between temperature and humidity after applying the DBSCAN clustering algorithm to remove outliers. The data points are grouped by weather conditions, revealing distinct clusters. For instance, high humidity values (above 80%) are associated with conditions like Rain, Overcast, while lower humidity values (below 60%) correspond to Clear conditions. This cleaning process enhances the clarity of patterns in the data.*

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

X = df[['temp', 'humidity']].dropna()
X_scaled = StandardScaler().fit_transform(X)

db = DBSCAN(eps=0.5, min_samples=5).fit(X_scaled)
df['dbscan_outlier'] = db.labels_

# -1 indicates noise/outliers
df_dbscan_clean = df[df['dbscan_outlier'] != -1]

# Visualize temp vs humidity
sns.scatterplot(data=df_dbscan_clean, x='temp', y='humidity', hue='conditions', alpha=0.6)
plt.title('DBSCAN Cleaned: Temp vs Humidity')
plt.show()
```
